```python
import torch
import torchvision
from torch.utils.tensorboard import SummaryWriter

writer = SummaryWriter("log")
resnet = torchvision.models.resnet18(pretrained=False)
resnet.conv1 = torch.nn.Conv2d(1, 64, kernel_size=7, stride=2, padding=3, bias=False)
transform = torchvision.transforms.Compose([torchvision.transforms.ToTensor(), torchvision.transforms.Normalize((0.5,), (0.5,))])
trainset = torchvision.datasets.MNIST('mnist_train', train=True, download=True, transform=transform)
trainloader = torch.utils.data.DataLoader(trainset, batch_size=64, shuffle=True)
images, labels = next(iter(trainloader))
```

```python
writer.add_graph(model=resnet, input_to_model=images, verbose=True)
```
